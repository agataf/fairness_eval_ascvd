```python
import pandas as pd
import os
import numpy as np
import logging
import sys
import torch
import copy
import yaml

import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
#sns.set_style("white")
sns.set_style("ticks")
custom_palette = ["red", "green", "orange", "blue"]
sns.set_palette(custom_palette)
grp_label_dict = {1: 'Black women', 2: 'White women', 3: 'Black men', 4: 'White men'}    

#EXPERIMENT_NAME = 'big_test_mmd_tuning'
#EXPERIMENT_NAME = 'big_test_erm_recalibration'
#EXPERIMENT_NAME = 'big_test_regularized_tuning'
EXPERIMENT_NAME = 'big_test_erm_tuning'

num_configs=1

args = {'experiment_name': EXPERIMENT_NAME,
        'cohort_path': '/labs/shahlab/projects/agataf/data/pooled_cohorts/cohort_extraction/all_cohorts.csv',
        'base_path': '/labs/shahlab/projects/agataf/data/cohorts/pooled_cohorts'
       }

output_path = '/labs/shahlab/projects/agataf/fairness_utility/eval/plots'
os.makedirs(output_path, exist_ok=True)
experiment_plot_path=os.path.join(output_path, 'big_test_erm_recalib')
os.makedirs(experiment_plot_path, exist_ok=True)
```

### get results

```python
standard_eval = []
outputs=[]
for config_id in range(num_configs):
    for fold_id in range(1,11):
        RESULT_PATH = os.path.join(args['base_path'], 'experiments', args['experiment_name'], 'performance',
                                           '.'.join((str(config_id), 'yaml')), str(fold_id))
        LOGGING_PATH = os.path.join(RESULT_PATH, 'training_log.log')
        
        CONFIG_PATH = os.path.join(args['base_path'], 'experiments', args['experiment_name'], 'config',
                                                   '.'.join((str(config_id), 'yaml')))


        config = yaml.load(open(CONFIG_PATH), Loader=yaml.SafeLoader)
        lambda_reg = config.get('lambda_group_regularization')

        result_df_group_standard_eval = (pd.read_parquet(
            os.path.join(RESULT_PATH,
                         'result_df_group_standard_eval.parquet'
                        ))
                                         .assign(fold_id=fold_id,
                                                 config_id=config_id,
                                                 lambda_reg=lambda_reg))
        standard_eval.append(result_df_group_standard_eval)
        
        output_df = (
            pd.read_parquet(
                os.path.join(RESULT_PATH,
                             'output_df.parquet'
                            )
            )
            .assign(fold_id=fold_id,
                    config_id=config_id,
                    lambda_reg=lambda_reg)
        )
        
        outputs.append(output_df)

df = pd.concat(standard_eval)
preds = pd.concat(outputs)

```

```python
preds0=preds.copy()
```

```python

aggregate_path = os.path.join(args['base_path'], 'experiments', 'big_test_erm_tuning', 'performance', 'all')
test_calibs = pd.read_csv(os.path.join(aggregate_path, 'predictions_test_recalibrated.csv'))
```

```python
test_calibs = (test_calibs
               .drop('pred_probs', axis=1)
               .rename(columns={'calibration_density': 'pred_probs'})
               .filter(preds.columns)
               .assign(calibrated=True)
              )
```

```python
preds = (preds
         .assign(calibrated=False)
         .query('phase=="test"')
         .append(test_calibs)
        )
```

### threshold-based performance metrics

```python
discovery_omission = pd.DataFrame(preds
                                  .query('(phase=="test")')
                                  .filter(['group', 'treat', 'person_id', 'lambda_reg', 'labels', 'fold_id'])
                                  .assign(f_discovery_rate_075 = lambda x: np.where((x.labels==0) & (x.treat>0), 1, 0),
                                          f_omission_rate_075 = lambda x: np.where((x.labels==1) & (x.treat==0), 1, 0),
                                          f_discovery_rate_20 = lambda x: np.where((x.labels==0) & (x.treat==2), 1, 0),
                                          f_omission_rate_20 = lambda x: np.where((x.labels==1) & (x.treat<2), 1, 0),
                                          fraction_treated = lambda x: np.where(x.treat>0, 1, 0),
                                         )
                                  .filter(['group', 'lambda_reg', 'fold_id',
                                           'f_discovery_rate_075', 'f_omission_rate_075',
                                           'f_discovery_rate_20', 'f_omission_rate_20',
                                           'fraction_treated'])
                                  .groupby(['lambda_reg', 'fold_id','group'])
                                  .mean()
                                  .stack()
                                  .reset_index()
                                  .rename(columns={'level_3':'metric', 0:'performance'})
                                 )
```

### get decision changes

```python
num_treated_by_stratum = (preds
     .query('(phase=="test") ')
     .filter(['group', 'treat', 'person_id', 'calibrated', 'labels'])
     .groupby(['calibrated',  'group', 'labels', 'treat'])
     .count()
    )

num_per_stratum = (preds
      .query('(phase=="test") ')

           .filter(['group', 'person_id', 'calibrated', 'labels'])
           .groupby(['calibrated',  'group', 'labels'])
           .count()
     )

frac_treated = num_treated_by_stratum.div(num_per_stratum)

frac_treated = (frac_treated
                .reset_index()
                .assign(group = lambda x: x.group.map(grp_label_dict),
                        labels = lambda x: x.labels.map({0: 'no ASCVD event',
                                                         1: 'ASCVD event'}))
               )
```

```python
frac_treated = frac_treated.assign(calibrated=lambda x: x.calibrated.astype(int))
```

```python
output_path = os.path.join(experiment_plot_path, 'prediction_distributions.png')


sns.despine(left=True)

sns.set(font_scale=1.5)
sns.set_style("ticks")
    
num_configs=2    
f, axes = plt.subplots(2,2, figsize=(23,10), sharex=False, sharey=False)
palette = iter(sns.color_palette("tab10",n_colors=2))
indices = {0: (0,0), 1: (1,0), 2: (0,1), 3: (1,1)}

labels_dict = {(0,0): 'ERM, no ASCVD',
              (0,1): 'ERM, ASCVD',
              (1,0): 'recalibrated, no ASCVD',
              (1,1): 'recalibrated, no ASCVD'}
for idx, grp in enumerate(range(1,5)):
    palette = iter(sns.color_palette("tab10",n_colors=2))

    for calibrated in [0,1]:
        color=next(palette)
#     for idx, grp in enumerate(range(1,5)):
        args = {'shade': False, 
                #'label': calibrated, 
                'ax': axes[indices[idx]], 
                'color':color}
        ax = sns.kdeplot(data=preds.query("calibrated==@calibrated & group==@grp & labels==0"), 
                         x="pred_probs",
                         **args,
                         label=labels_dict.get(calibrated, 0),
                         legend=True)
        ax = sns.kdeplot(data=preds.query("calibrated==@calibrated & group==@grp & labels==1"), 
                         x="pred_probs",
                         **args,
                         label=labels_dict.get(calibrated, 1),
                         linestyle="--",
                        legend=True)

        
        ax.set(xlim=(-0.1,1))
        ax.set(ylim=(0,17.5))
        ax.set_title(grp_label_dict[grp])
        ax.set_xlabel("")
        ax.axvline(x=0.075, linestyle='--', color='grey')
        ax.axvline(x=0.2, linestyle='--', color='grey')
        
sns.set(font_scale=1)
ax.legend(title='stratum (0=no outcome, 1=outcome) \n blue=ERM, orange=recalibrated', fontsize=15, bbox_to_anchor=(0.5, 1),
         title_fontsize=15)

if output_path is not None:
    f.savefig(output_path)
```
